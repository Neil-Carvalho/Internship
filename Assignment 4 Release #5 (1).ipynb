```python
#14/01
url = 'https://github.com/'
driver.get(url) #Loading the Webpage
driver.implicitly_wait(4) #Waiting for the page to load

#Navigating to the requrired webpage
#driver.get(driver.find_element_by_xpath("//a[@href='/trending']").get_attribute("href"))
driver.get(driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul/li[5]/a").get_attribute("href"))
driver.implicitly_wait(4)

#Declaring the variables
Repo_Name_List = []
Repo_Description_List = []
Repo_Language_List = []
Number_Of_Contributors = []
Repo_Links_List = []
Repo_ContributorCount = []

#Populating the variable
for name in driver.find_elements_by_xpath("//h1[1]/a[1]"):
    Repo_Name_List.append(name.text) 
    

count_of_repositories_on_page = len(driver.find_elements_by_xpath("//h1[1]/a[1]"))
for dpe in range(0,count_of_repositories_on_page):
    LanguageUsedElement = '//article[' + str(dpe+1) + ']/div[2]/span[1]/span[2]'
    try:
        Repo_Language_List.append(driver.find_element_by_xpath(LanguageUsedElement).text)
    except:
        Repo_Language_List.append("-")

for link in driver.find_elements_by_xpath("//h1[1]/a[1]"):
    Repo_Links_List.append(link.get_attribute("href"))

for url in Repo_Links_List:
    driver.get(url)
    driver.implicitly_wait(4)
    try:
        Repo_ContributorCount.append(driver.find_element_by_xpath("//a[contains(text(),'Contributors')]//span").text)
    except:
        Repo_ContributorCount.append('N/A')
    
    try:
        Repo_Description_List.append(driver.find_element_by_xpath("//div[@class = 'BorderGrid-cell']/p").text)
    except:
        Repo_Description_List.append('N/A')
        

#declaring the DataFrame
df = pd.DataFrame({'Repository Titile':Repo_Name_List, 
                   'Repository Description': Repo_Description_List,
                   'Language Used': Repo_Language_List,
                   'Contributor Count': Repo_ContributorCount})
#Displaying the DataFrame
df
```

```python
#14/01
url = 'https://www.imdb.com/list/ls095964455/' #Loading the website
driver.get(url) #loading the website
driver.implicitly_wait(4) # waiting

#Initialinsing the variables
Name_List = []
Year_Span_List = []
Genre_List = []
Run_time_List = []
Ratings_List = []
Votes_List = []

#Populating the variables
for Series_Names in driver.find_elements_by_xpath("//div[2]/h3/a"):
    Name_List.append(Series_Names.text)
    
for Year_Span in driver.find_elements_by_xpath("//div[2]/h3/span[2]"):
    Year_Span_List.append(Year_Span.text)

for Genre in driver.find_elements_by_xpath("//div[2]/p[1]/span[5]"):
    Genre_List.append(Genre.text)
    
for Run_time in driver.find_elements_by_xpath("//div[2]/p[1]/span[3]"):
    Run_time_List.append(Run_time.text.split(' ')[0])

for Ratings in driver.find_elements_by_xpath("//div/div/span[@class = 'ipl-rating-star__star']//..//span[2]"):
    Ratings_List.append(Ratings.text)
    
for Votes in driver.find_elements_by_xpath("//div[2]/p[4]/span[2]"):
    Votes_List.append(Votes.text)

#Checking to confirm all elements are of the same length
print(len(Name_List))
print(len(Year_Span_List))
print(len(Genre_List))
print(len(Run_time_List))
print(len(Ratings_List))
print(len(Votes_List))

#creating the dataframe
df = pd.DataFrame({'Series Name': Name_List, 'Year Span': Year_Span_List,'Genre':Genre_List, 
                   'Run Time(in Minutes)':Run_time_List,'Ratings': Ratings_List,'Vote Count':Votes_List})
#displaying the dataframe
df
```

```python
#14/01
#print("started")
url = 'https://archive.ics.uci.edu/ml/index.php'
driver.get(url)
driver.implicitly_wait(4)
driver.get(driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b//..//..").get_attribute("href"))
#print(driver.find_element_by_xpath("").get_attribute("href"))

Dataset_Name_List = []
Data_Type_List = []
Task_List = []
Attribute_Type_List = []
No_Of_Instances_List = []
No_Of_Attributes_List = []
Year_List = []

Dataset_Name_List = []
Data_Type_List = []   
Task_List = []        
Attribute_Type_List = []
No_Of_Instances_List = []
No_Of_Attributes_List = []
Year_List = []            

for name in driver.find_elements_by_xpath("//table[2]//b/a"):
    Dataset_Name_List.append(name.text)
    
for tasktype in driver.find_elements_by_xpath("//table[2]//td[2]//td[3]/p"):
    Task_List.append(tasktype.text)

for attributetype in driver.find_elements_by_xpath("//table[2]//td[2]//td[4]/p"):
    Attribute_Type_List.append(attributetype.text)
    
for instancecount in driver.find_elements_by_xpath("//table[2]//td[2]//td[5]/p"):
    No_Of_Instances_List.append(instancecount.text)

for attributecount in driver.find_elements_by_xpath("//table[2]//td[2]//td[6]/p"):
    No_Of_Attributes_List.append(attributecount.text)

for YearList in driver.find_elements_by_xpath("//table[2]//td[2]//td[7]/p"):
    Year_List.append(YearList.text)

for i in range(len(driver.find_elements_by_xpath("//table[2]//b/a"))):
    temp_path = '//tr[' + str(i + 2) + ']/td[2]/p[1]'
    Data_Type_List.append(driver.find_element_by_xpath(temp_path).text)

Task_List.pop(0)
Attribute_Type_List.pop(0)
No_Of_Instances_List.pop(0)
No_Of_Attributes_List.pop(0)
Year_List.pop(0)

print(len(Dataset_Name_List))
print(len(Data_Type_List))
print(len(Task_List))
print(len(Attribute_Type_List))
print(len(No_Of_Instances_List))
print(len(No_Of_Attributes_List))
print(len(Year_List))


df = pd.DataFrame({'Dataset Name': Dataset_Name_List, 'Data Type': Data_Type_List,'Task List': Task_List,
                   'Attribute Type': Attribute_Type_List, 'No Of Instances': No_Of_Instances_List,
                   'No_Of_Attributes_List':No_Of_Attributes_List,'Year Created': Year_List})
df
```
